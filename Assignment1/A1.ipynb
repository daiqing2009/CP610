### Question 1. Basic understanding about the distribution of the data. For nominal attributes, what are frequencies of the possible states? For numerical attributes, what are the statistics (min, max, median, quantiles)? 

```python
# frequencies of possible states of each nominal attributes
for column in nom_attrs:
    print(column + ':')
    print(df[column].value_counts().to_dict())

# statistics(min, max, median, quantiles) of 
def custom_describe(data):
    description = data.describe()
    # min, max, quantiles already included in the describe() function only add median to it. 
    median = data.median()    
    description.loc['median'] = median 
    return description

print(custom_describe(df[num_attrs]))
```

### Question 2. Visualization of the nominal data. For the attribute "home_team", plot its states and corresponding frequencies in (1) a bar chart, and (2) a dot plot. 

```python
import matplotlib.pyplot as plt

# Count the frequency of each state in the "home_team" column
state_counts = df['home_team'].value_counts()

# print(state_counts.head())

# Create a bar chart
plt.figure(figsize=(12, 6))
state_counts.plot(kind='bar',color='skyblue')
plt.title('Frequency of Home Teams (Bar Chart)')
plt.xlabel('Home Team')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Create a dot plot
plt.figure(figsize=(12, 6))
plt.plot(state_counts,'o')
plt.title('Frequency of Home Teams (Dot Plot)')
plt.xlabel('Home Team')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Question 4. Which features have missing values? Please list them. Fill in the missing values using the mode for nominal attributes and the median for numerical attributes. After the imputation, check the number of missing values again.

```python
#check the missing values in the attributes:
missing_values = df[nom_attrs].isna().sum()
print("Missing values in nomical attribute(before fill in):")
print(missing_values)

# Calculate mode of nominal attributes 
mode= df[nom_attrs].mode(dropna=True).iloc[0]
print("mode of nominal attributes:" )
print(mode)

# fill the missing values with mode for nominal attributes 
for column in nom_attrs:
    if df[column].isna().any():  # Check if column has missing values
        print('Empty value of column['+column+'] will be filled with:' + mode[column])
        # Fill missing values in the column with mode
        df[column].fillna(mode[column], inplace=True)

#check the missing values in the attributes:
missing_values = df[nom_attrs].isna().sum()
print("Missing values in nomical attribute(after fill in):")
print(missing_values)
```

```python
#check the missing values in the attributes:
missing_values = df[num_attrs].isna().sum()
print("\nMissing values in numberical attribute(before fill in):")
print(missing_values)

# Calculate median of numberical attributes 
median= df[num_attrs].median(skipna=True)
print("median of numberical attributes:" )
print(median)
# fill the missing values with median for numerical attributes 
for column in num_attrs:
    if df[column].isna().any():  # Check if column has missing values
        print('Empty value of column['+column+'] will be filled with:' + str(median[column]))
        # Fill missing values in the column with mode
        df[column].fillna(median[column], inplace=True)

#check the missing values in the attributes:
missing_values = df[num_attrs].isna().sum()
print("\nMissing values in numberical attribute(after fill in):")
print(missing_values)
```

* fill in the missing value of nominal data with mode 
* fill in the missing value of numerical data with the median

* posteam/defteam will not necessarily be the most frequent values, far from that actualy, the two components are very likely to be Neither of the most frequent posteam/defteam!
* When an event happens, it can be of any EP, considering the wide distribution of EP(refer to the explanation in the next paragraph)
